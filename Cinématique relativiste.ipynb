```python
#Facteur de Lorentz

c = 3*10**8 
v = np.linspace (0.5*c, 0.99995*c, 1000)

def gamma(x) : 
    return 1/np.sqrt(1-x**2)

v_muons = 0.995*c
gamma_muons = np.ones(1000)*gamma(v_muons/c)

plt.figure(9,figsize=(9, 9))
plt.plot(v/c, gamma(v/c))
plt.plot(v/c, gamma_muons,'red', linewidth = 1, label='v_muons = 0.995*c  -> $ \gamma \sim$ 10')
plt.grid(which='both')
plt.xlabel('v/c', fontsize=18)
plt.ylabel('$\gamma$', fontsize=18)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.legend()
plt.legend(fontsize=20)
plt.show()
```

```python
#Population des muons

tau = 2.2*10**-6

def N(x) : 
    return np.exp(-x/(c*tau))

distance = np.linspace(0, 15*10**3, 1000)

plt.figure(19,figsize=(9, 9))
plt.plot(distance, N(distance))
plt.plot(distance, N(distance/gamma(v_muons/c)),'red', linewidth = 1, label='v_muons = 0.995*c  -> $\gamma \sim$ 10')
plt.grid(which='both')
plt.xlabel('distance parcourue (m)', fontsize=18)
plt.yscale('log')
plt.ylabel('N/N0', fontsize=18)
#plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.legend()
plt.legend(fontsize=20)
plt.show()
```
