```python
import numpy as np
import math, cmath
from matplotlib.pyplot import *
from scipy.constants import *
i = complex(0,1)
```

```python
                                    ###### PARTIE REELLE DE LA FONCTION D'ONDE ########


#Creation de l'abscisse

def abscisse (xmin, xmax, points = 50000) : 

    return np.linspace(xmin, xmax, points)

##Estimation de la periode afin de calibrer la feneetre des abscisses

##Tracé

def r(e,L) : 
    t = cmath.sqrt(e/(1-e)) #t = k/K
    t_plus = 1+i*t
    t_moins = 1-i*t
    u = 2*cmath.pi*cmath.sqrt(e)*L #ka
    v = 2*cmath.pi*cmath.sqrt(1-e)*L #Ka
    r = (-2*cmath.sinh(v)*t_plus*t_moins)/(t_moins**2*cmath.exp(v)-cmath.exp(-v)*t_plus**2)
    return r

def t(e,L) : 
    t = cmath.sqrt(e/(1-e)) #t = k/K
    t_plus = 1+i*t
    t_moins = 1-i*t
    u = 2*cmath.pi*cmath.sqrt(e)*L #ka
    v = 2*cmath.pi*cmath.sqrt(1-e)*L #Ka
    tau = (cmath.exp(-i*u)*(t_moins**2-t_plus**2))/(t_moins**2*cmath.exp(v)-cmath.exp(-v)*t_plus**2)
    return tau

def Phi(e,L) : #e=E/V0 ; L = a/lambda_0
    X = abscisse(-4/(L*e**0.5), 0) #X/a
    Y = []
    Z = []
    ref = r(e,L)
    periode = [1+ref]
    for k in range (len(X)) : 
        Y.append(cmath.exp(i*2*cmath.pi*cmath.sqrt(e)*L*X[k])+ref*cmath.exp(-i*2*cmath.pi*cmath.sqrt(e)*L*X[k]))
        Z.append(Y[k].real)
    plot(X,Z)

def Phi2(e,L) : #e=E/V0 ; L = a/lambda_0
    t = cmath.sqrt(e/(1-e)) #t = k/K
    X = abscisse(0,1)
    Y = []
    Z = []
    ref = r(e,L)
    B_2 = 0.5*(1+i*t+(1-i*t)*ref)
    B_3 = 0.5*(1-i*t+(1+i*t)*ref)

    for k in range (len(X)) : 
        Y.append(B_2*cmath.exp(2*cmath.pi*cmath.sqrt(1-e)*L*X[k])+B_3*cmath.exp(-2*cmath.pi*cmath.sqrt(1-e)*L*X[k]))
        Z.append(Y[k].real)
    plot(X,Z)

def Phi3(e,L) : #e=E/V0 ; L = a/lambda_0
    X = abscisse(1,1+4/(L*e**0.5))
    Y = []
    Z = []
    tau = t(e,L)

    for k in range (len(X)) : 
        Y.append(tau*cmath.exp(i*2*cmath.pi*cmath.sqrt(e)*L*X[k]))
        Z.append(Y[k].real)
    plot(X,Z)

def Phi_module(e,L) : #e=E/V0 ; L = a/lambda_0
    X = abscisse(-4/(L*e**0.5), 0) #X/a
    Y = []
    Z = []
    ref = r(e,L)
    periode = [1+ref]
    for k in range (len(X)) : 
        Y.append(cmath.exp(i*2*cmath.pi*cmath.sqrt(e)*L*X[k])+ref*cmath.exp(-i*2*cmath.pi*cmath.sqrt(e)*L*X[k]))
        Z.append(abs(Y[k])**2)
    plot(X,Z)

def Phi2_module(e,L) : #e=E/V0 ; L = a/lambda_0
    t = cmath.sqrt(e/(1-e)) #t = k/K
    X = abscisse(0,1)
    Y = []
    Z = []
    ref = r(e,L)
    B_2 = 0.5*(1+i*t+(1-i*t)*ref)
    B_3 = 0.5*(1-i*t+(1+i*t)*ref)

    for k in range (len(X)) : 
        Y.append(B_2*cmath.exp(2*cmath.pi*cmath.sqrt(1-e)*L*X[k])+B_3*cmath.exp(-2*cmath.pi*cmath.sqrt(1-e)*L*X[k]))
        Z.append(abs(Y[k])**2)
    plot(X,Z)

def Phi3_module(e,L) : #e=E/V0 ; L = a/lambda_0
    X = abscisse(1,1+4/(L*e**0.5))
    Y = []
    Z = []
    tau = t(e,L)

    for k in range (len(X)) : 
        Y.append(tau*cmath.exp(i*2*cmath.pi*cmath.sqrt(e)*L*X[k]))
        Z.append(abs(Y[k])**2)
    plot(X,Z)

def Effet_tunnel_Re (e,L) : 
    var = (e,L)
    Phi(e,L)
    Phi2(e,L)
    Phi3(e,L)
    xlabel(r"$x/a$")
    ylabel(r"$Re(\Phi)$")
    title ("(e,L) = {}".format(var))
    show()

def Effet_tunnel_module (e,L) : 
    var = (e,L)
    Phi_module(e,L)
    Phi2_module(e,L)
    Phi3_module(e,L)
    xlabel(r"$x/a$")
    ylabel(r"$Module carré de Phi$")
    title ("(e,L) = {}".format(var))
    show()

Effet_tunnel_Re(0.5, 0.5)
```
